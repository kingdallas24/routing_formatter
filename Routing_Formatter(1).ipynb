```python
def extract_zip(row):
    if "*" in row['ZIP']:
        try:
            # Geocode the city to get latitude and longitude
            location = nom.geocode(row['City'])
            if location:
                # Reverse geocode to get the address and extract the zip code
                reverse_location = nom.reverse((location.latitude, location.longitude))
                zip_code = None
                for component in reverse_location.raw.get("address", []):
                    if component.get("postcode"):
                        zip_code = component.get("postcode")
                        break
                if zip_code:
                    return zip_code
            return None
        except Exception as e:
            print(f"Error: {e}")
            return None
    else:
        return row['ZIP']

df2['ZIP'] = df2.apply(extract_zip, axis=1)
```

# Get ZIP from city name if ZIP is Incorrect

```python
def get_state_from_zip(zip):
    # Create a Nominatim geolocator
    
    try:
        # Geocode the address
        location = nom.geocode(zip)
        
        # Check if a location was found
        if location:
            # Split the address by commas and extract the state information
            result_list = location.address.split(", ")
            state = result_list[-1]  # Assuming the state is the second-to-last element
            
            # Remove leading and trailing whitespace
            state = state.strip()
            
            return state
        else:
            return "Location not found"
    except Exception as e:
        return str(e)
```

```python
def get_city_from_zip(zip):
    # Create a Nominatim geolocator
    
    try:
        # Geocode the address
        location = nom.geocode(zip)
        
        # Check if a location was found
        if location:
            # Split the address by commas and extract the state information
            result_list = location.address.split(", ")
            state = result_list[-2]  # Assuming the state is the second-to-last element
            
            # Remove leading and trailing whitespace
            state = state.strip()
            
            return state
        else:
            return "Location not found"
    except Exception as e:
        return str(e)
```

```python
# Define a function to create the full address
def create_full_address(row):
    # Check if 'Unique_Roads' is not NaN
    if not pd.isna(row['Unique_Road']) and len(row['Unique_Road']) > 5:
        # Check if a comma already exists in Unique_Roads
        if ',' in row['Unique_Road']:
            # If there's a comma, just concatenate the columns
            full_address = f"{row['Unique_Road']} {row['City']}, {row['State']} {row['ZIP']}"
        else:
            # If no comma exists, add one before concatenating
            full_address = f"{row['Unique_Road']}, {row['City']}, {row['State']} {str(row['ZIP'])}"
    else:
        # Handle cases where 'Unique_Roads' is NaN
        full_address = f"{row['City']}, {row['State']} {str(row['ZIP'])}"
    
    return full_address

# Apply the function to create the full_address column
df2['full_address'] = df2.apply(create_full_address, axis=1)
```
